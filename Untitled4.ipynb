```python
# 🛠️ Install required libraries
!pip install transformers pillow torch torchvision --quiet

# 📌 Importing libraries

from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# 📌 Load a pre-trained BLIP model (Vision + Language)

print("🔄 Loading the Image Captioning model ...")
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# 📌 Upload your own image

from google.colab import files

print("📤 Please upload an image...")
uploaded = files.upload()

image_path = list(uploaded.keys())[0]  # get the uploaded file name
image = Image.open(image_path).convert("RGB")

# 📌 Generate a caption

print("🖼️ Understanding your image...")
inputs = processor(image, return_tensors="pt")

with torch.no_grad():
    output = model.generate(**inputs, max_length=30)

caption = processor.decode(output[0], skip_special_tokens=True)

# 📌 Show result

print("\n✅ Here’s a friendly caption for your image:")
print(f"👉 {caption}")
```

```python
# -----------------------------------------------------
# 📌 Image Captioning AI (with Image Display)
# -----------------------------------------------------
# This script will:
# 1. Let you upload an image
# 2. Show the image inside Colab
# 3. Generate a friendly caption for it
# -----------------------------------------------------

# 🛠️ Install required libraries
!pip install transformers pillow torch torchvision --quiet

# -----------------------------------------------------
# 📌 Importing libraries
# -----------------------------------------------------
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------------------------------
# 📌 Load a pre-trained BLIP model (Vision + Language)
# -----------------------------------------------------
print("🔄 Loading the Image Captioning model ...")
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# -----------------------------------------------------
# 📌 Upload your own image
# -----------------------------------------------------
print("📤 Please upload an image...")
uploaded = files.upload()

image_path = list(uploaded.keys())[0]  # get uploaded file name
image = Image.open(image_path).convert("RGB")

# -----------------------------------------------------
# 📌 Display the uploaded image
# -----------------------------------------------------
plt.imshow(image)
plt.axis("off")
plt.title("🖼️ Uploaded Image", fontsize=14)
plt.show()

# -----------------------------------------------------
# 📌 Generate a caption
# -----------------------------------------------------
print("🖼️ Understanding your image...")
inputs = processor(image, return_tensors="pt")

with torch.no_grad():
    output = model.generate(**inputs, max_length=30)

caption = processor.decode(output[0], skip_special_tokens=True)

# -----------------------------------------------------
# 📌 Show result
# -----------------------------------------------------
print("\n✅ Here’s a friendly caption for your image:")
print(f"👉 {caption}")
```
