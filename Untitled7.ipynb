```python

# -----------------------------------------------------
# 🤖 Humanized Image Captioning AI (Loop Version)
# -----------------------------------------------------
# This script will:
# 1. Let you upload as many images as you like
# 2. Display each image
# 3. Generate a nice caption
# 4. Keep asking until you say "stop"
# -----------------------------------------------------

# 🛠️ Install required libraries
!pip install transformers pillow torch torchvision --quiet

# -----------------------------------------------------
# 📌 Importing libraries
# -----------------------------------------------------
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------------------------------
# 📌 Load the BLIP model (Vision + Language)
# -----------------------------------------------------
print("🔄 Loading the Image Captioning model ... please wait...")
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
print("✅ Model loaded successfully!")

# -----------------------------------------------------
# 📌 Friendly Captioning Loop
# -----------------------------------------------------
while True:
    print("\n📤 Please upload an image (or type 'stop' to exit):")
    user_input = input("👉 Type 'stop' to quit, or press Enter to upload: ").strip().lower()
    if user_input == "stop":
        print("👋 Thank you for using the Image Captioning AI! Goodbye!")
        break

    # Upload an image
    uploaded = files.upload()
    image_path = list(uploaded.keys())[0]
    image = Image.open(image_path).convert("RGB")

    # Show the uploaded image
    plt.imshow(image)
    plt.axis("off")
    plt.title("🖼️ Your Uploaded Image", fontsize=14)
    plt.show()

    # Generate a caption
    print("✨ Thinking about your image...")
    inputs = processor(image, return_tensors="pt")

    with torch.no_grad():
        output = model.generate(**inputs, max_length=30)

    caption = processor.decode(output[0], skip_special_tokens=True)

    # Show result
    print("\n✅ Here’s what I think about your image:")
    print(f"👉 {caption}")
```
