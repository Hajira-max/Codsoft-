```python
# --------------------------------------------------
# 📌 Image Captioning Project (GitHub-safe)
# --------------------------------------------------
# Features:
# 1. Upload an image, see AI-generated caption.
# 2. After each caption, choose Yes/No to continue.
# 3. Image is displayed with its caption.
# 4. At the end, you can clean & download a copy of
#    the notebook that works perfectly on GitHub.
# --------------------------------------------------

# Step 1: Install libraries
!pip install transformers sentencepiece --quiet

# Step 2: Import tools
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer
import torch
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files
import json
import os

# Step 3: Load the pre-trained model
model_name = "nlpconnect/vit-gpt2-image-captioning"
print("📥 Loading model... please wait a moment.")
model = VisionEncoderDecoderModel.from_pretrained(model_name)
feature_extractor = ViTImageProcessor.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print("✅ Model ready!")

# Step 4: Function to generate caption
def generate_caption(image_path):
    img = Image.open(image_path).convert("RGB")
    pixel_values = feature_extractor(images=img, return_tensors="pt").pixel_values.to(device)
    output_ids = model.generate(pixel_values, max_length=16, num_beams=4)
    caption = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Show image with caption
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"📝 Caption: {caption}", fontsize=12, color="red")
    plt.show()

    return caption

# Step 5: Loop with Yes/No option
while True:
    print("\n📂 Please upload an image (JPG/PNG).")
    uploaded = files.upload()

    for filename in uploaded.keys():
        print("--------------------------------------------------")
        print(f"📷 Image: {filename}")
        caption = generate_caption(filename)
        print(f"📝 Caption: {caption}")

    # Ask user if they want to continue
    choice = input("\n👉 Do you want to upload another image? (yes/no): ").strip().lower()
    if choice != "yes":
        print("👋 Exiting captioning loop.")
        break

# --------------------------------------------------
# Step 6: Clean notebook before saving to GitHub
# --------------------------------------------------

def clean_notebook(input_path, output_path):
    """
    Remove widget metadata so GitHub renders the notebook correctly.
    """
    with open(input_path, "r", encoding="utf-8") as f:
        nb = json.load(f)

    if "widgets" in nb.get("metadata", {}):
        nb["metadata"].pop("widgets", None)

    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(nb, f, indent=2)

    print(f"✅ Cleaned notebook saved as: {output_path}")

# Automatically detect current notebook name (fallback to 'Image_Captioning.ipynb')
notebook_name = os.environ.get("COLAB_NOTEBOOK_NAME", "Image_Captioning.ipynb")
clean_name = notebook_name.replace(".ipynb", "_clean.ipynb")

# Clean + Save
try:
    clean_notebook(notebook_name, clean_name)
    files.download(clean_name)
except Exception as e:
    print("⚠️ Could not clean automatically inside Colab.")
    print("👉 To fix: Save notebook as .ipynb from Colab, then run the clean_notebook() function manually.")
```
